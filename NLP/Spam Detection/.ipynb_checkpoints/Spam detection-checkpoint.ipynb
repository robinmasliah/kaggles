```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.get_backend()
import seaborn as sns
plt.style.use('seaborn')
import re
import string
import math
import ast
from sklearn import preprocessing
import textstat
from utils import *
# Importing spacy
import spacy
# Loading model
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
df = pd.read_csv('spam_text.csv')
```

```python
df = df[['Category','Message']]
print("Shape of data => ", df.shape)
df.head(3)
```

```python
df.isnull().sum()
```

```python
# Dictionary of English Contractions
with open('dict.txt') as f:
    contractions_dict = f.read()
contractions_dict = ast.literal_eval(contractions_dict)

# Regular expression for finding contractions
contractions_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))

# Function for expanding contractions
def expand_contractions(text,contractions_dict=contractions_dict):
    def replace(match):
        return contractions_dict[match.group(0)]
    return contractions_re.sub(replace, text)

df['Message'] = df['Message'].apply(lambda x: x.lower())
# Expanding Contractions in the mails
df['Message'] = df['Message'].apply(lambda x: expand_contractions(x))
df['Message'] = df['Message'].apply(lambda x: re.sub('\w*\d\w*','', x))
df['Message'] = df['Message'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
# Removing extra spaces
df['Message'] = df['Message'].apply(lambda x: re.sub(' +',' ',x))
# Lemmatization with stopwords removal
df['lemmatized'] = df['Message'].apply(lambda x: ' '.join([token.lemma_ for token in list(nlp(x)) if (token.is_stop==False)]))
df_grouped = df[['Category', 'lemmatized']].groupby(by='Category').agg(lambda x:' '.join(x))
```

```python
# Creating Document Term Matrix
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(analyzer='word')
data = cv.fit_transform(df_grouped['lemmatized'])
df_dtm = pd.DataFrame(data.toarray(), columns=cv.get_feature_names())
df_dtm.index = df_grouped.index
```

```python
# WordCloud
# Transposing document term matrix
df_dtm = df_dtm.transpose()

# Plotting word cloud for each product
for index,product in enumerate(df_dtm.columns):
    generate_wordcloud(df_dtm[product].sort_values(ascending=False),product)
```

```python
from textblob import TextBlob
df['polarity'] = df['lemmatized'].apply(lambda x:TextBlob(x).sentiment.polarity)
```

```python
print("3 Random mails with Highest Polarity:")
for index,review in enumerate(df.iloc[df['polarity'].sort_values(ascending=False)[:3].index]['Message']):
    print('mail {}:\n'.format(index+1),review)
```

```python
print("3 Random mails with Lowest Polarity:")
for index,review in enumerate(df.iloc[df['polarity'].sort_values(ascending=True)[:3].index]['Message']):
    print('mail {}:\n'.format(index+1),review)
```

```python
product_polarity_sorted = pd.DataFrame(df.groupby('Category')['polarity'].mean().sort_values(ascending=True))

plt.figure(figsize=(16,8))
plt.xlabel('Polarity')
plt.ylabel('Products')
plt.title('Polarity of Spams')
polarity_graph=plt.barh(np.arange(len(product_polarity_sorted.index)),
                        product_polarity_sorted['polarity'],
                        color='purple',)

# Writing product names on bar
for bar,product in zip(polarity_graph,product_polarity_sorted.index):
    plt.text(0.005, bar.get_y()+bar.get_width(),'{}'.format(product), 
             va='center', fontsize=11, color='white')

# Writing polarity values on graph
for bar,polarity in zip(polarity_graph,product_polarity_sorted['polarity']):
    plt.text(bar.get_width() + 0.001,bar.get_y() + bar.get_width(),
             '%.3f'%polarity, 
             va='center',
             fontsize=11,color='black')

    plt.yticks([])
plt.show()
```

```python
df['dale_chall_score'] = df['Message'].apply(lambda x: textstat.dale_chall_readability_score(x))
df['flesh_reading_ease'] = df['Message'].apply(lambda x: textstat.flesch_reading_ease(x))
df['gunning_fog'] = df['Message'].apply(lambda x: textstat.gunning_fog(x))
df['text_standard'] = df['Message'].apply(lambda x: textstat.text_standard(x))
df['reading_time'] = df['Message'].apply(lambda x: textstat.reading_time(x))
```

```python
le = preprocessing.LabelEncoder()
le.fit(df.Category)
df.Category = le.transform(df.Category)

le.fit(df.text_standard)
df.text_standard = le.transform(df.text_standard)
```

```python
import multiprocessing

from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import nltk

from matplotlib import pyplot as plt
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
corpus = []
for col in df.Message:
    word_list = col.split(" ")
    corpus.append(word_list)

#generate vectors from corpus
model = Word2Vec(min_count=20, window=2, size=300, sample=6e-5, alpha=0.03, min_alpha=0.0007, negative=20)

model.build_vocab(corpus, progress_per=10000)
```

```python
#list the vocabulary words
words = list(model.wv.vocab)
#or show the dictionary of vocab words
#model.wv.vocab
```

```python
#pass the embeddings to PCA
X = model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
#create df from the pca results
pca_df = pd.DataFrame(result, columns=['x', 'y'])
#add the words for the hover effect
pca_df['word'] = words
```

```python
X = model[model.wv.vocab]
pca = PCA(n_components=2)

result = pca.fit_transform(X)

# create a scatter plot of the projection
plt.scatter(result[:, 0], result[:, 1])
words = list(model.wv.vocab)

for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))

plt.show()
```
