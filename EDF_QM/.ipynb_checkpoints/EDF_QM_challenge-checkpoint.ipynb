```python
# Fonction pour dissocier la date de l'heure
def split_it_space(year):
    return re.findall('([^\s]+)', year)
```

```python
# Tranformation de la colonne datetime en deux colonnes distinctes heure et date
meteo_train['Date UTC'] = meteo_train['Date UTC'].apply(lambda x: split_it_space(x))
```

```python
meteo_train['date'] = 0
meteo_train['heure'] = 0
```

```python
for i in range(0, len(meteo_train['Date UTC'])):
    meteo_train['date'][i] = meteo_train['Date UTC'][i][0]
```

```python
for i in range(0, len(meteo_train['Date UTC'])):
    meteo_train['heure'][i] = meteo_train['Date UTC'][i][1]
```

```python
del meteo_train['Date UTC']
```

```python
# Création d'un index date
meteo_train["date"] = pd.to_datetime(meteo_train["date"], errors="ignore", format="%d/%m/%Y")
meteo_train.set_index('date', inplace=True)
```
