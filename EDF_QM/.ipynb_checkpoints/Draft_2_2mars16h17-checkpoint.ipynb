```python
#load data
conso_train = pd.read_csv('conso_train.csv', sep=';', parse_dates=True)
meteo_train = pd.read_csv('meteo_train.csv', sep=';', parse_dates=True)
sample_solution = pd.read_csv('sample_solution.csv', sep=';', parse_dates=True, header=None)
meteo_prev = pd.read_csv('meteo_prev.csv', sep=';', parse_dates=True)
```

```python
meteo_train["Date UTC"] = pd.to_datetime(meteo_train["Date UTC"])
meteo_train()
```

```python
meteo_train['date'] = meteo_train['Date UTC'].str.split(" ").apply(lambda x:x[0])
```

```python
meteo_train['heure'] = meteo_train['Date UTC'].str.split(" ").apply(lambda x:x[1])
```

Aggreger les conso par pas de 3h pour qu'elles soient sur le même pas que les séries temporelles

```python
#traitement de la date conso_train
conso_train["date"] = conso_train["date"].str.split('+').apply(lambda x:x[0])
```

```python
conso_train["date_only"] = conso_train["date"].str.split('T').apply(lambda x:x[0])
conso_train["heure_only"] = conso_train["date"].str.split('T').apply(lambda x:x[1])                                                                
```

```python
conso_train["full_date"] = conso_train["date"].str.replace('T',' ')
conso_train.drop('date',axis=1)
```

```python
conso_train["full_date"] = pd.to_datetime(conso_train["full_date"], format="%Y-%m-%d %H:%M:%S")
```

```python
conso_train["full_date"] = pd.to_datetime(conso_train["full_date"], format="%Y-%m-%d %H:%M:%S")
```

```python
conso_train.set_index('full_date', inplace=True)
conso_train.head()
```

```python
power = conso_train['2015-09-13':'2016-09-13'].groupby(pd.Grouper(freq='3H')).sum()
```

```python
# Drop l'index et le remplace par un index numérique par défaut
power = power.reset_index()
```

```python
len(power)
```

```python
train = pd.concat([meteo_train, power], axis=1)
```

```python
del train['Date UTC']
del train['full_date']
```

```python
train['date_info'] = train['date']
```

```python
#changer le format d'une date en datetime
train["date"] = pd.to_datetime(train.date)
train["date_info"] = pd.to_datetime(train.date)
```

```python
train["date"] = pd.to_datetime(train["date"], format="%Y-%m-%d")
train.set_index('date', inplace=True)
train.tail()
```

```python
train = train.append([train]*2,ignore_index=True)
train.sort_values(by='date_info', axis=0, ascending=False)
```

```python
df_ts = train['puissance']
# Création des moyennes hebdomadaires
df_ts = df_ts.resample('M').mean()
df_ts = df_ts.dropna()
# Année 2011
ax = df_ts.plot()
ax.set_xlim(pd.Timestamp('2015-09-13'), pd.Timestamp('2015-12-30'))

plt.xlabel('Date')
plt.ylabel('Consommation')
plt.title("Consommation moyenne par mois en 2015")

plt.axis()
plt.legend()
plt.show()
```

```python
# Etude des corrélations entre les features
f, ax = plt.subplots(figsize=(10, 8))
corr = train.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
# SQL engine pour pouvoir explorer les données à l'oeil nu rapidement à titre observatoire.
engine = create_engine('sqlite:///:memory:')
train.to_sql('table', engine)
pd.read_sql_table('table', engine)
pd.read_sql_query('select * from "table" limit 5', con=engine)
```
