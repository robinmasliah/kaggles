```python
#load data
conso_train = pd.read_csv('conso_train.csv', sep=';', parse_dates=True)
meteo_train = pd.read_csv('meteo_train.csv', sep=';', parse_dates=True)
sample_solution = pd.read_csv('sample_solution.csv', sep=';', parse_dates=True,header=None)
meteo_prev = pd.read_csv('meteo_prev.csv', sep=';', parse_dates=True)
```

objectif : split date & heure

   Split date et heure méthode 2 : revoir .str , apply()_ Cours migli, to_datetime et date index

```python
meteo_train['date'] = meteo_train['Date UTC'].str.split(" ").apply(lambda x:x[0])
```

```python
meteo_train['heure'] = meteo_train['Date UTC'].str.split(" ").apply(lambda x:x[1])
```

```python
meteo_train.head()
```

Aggreger les conso par pas de 3h pour qu'elles soient sur le même pas que les séries temporelles

```python
#traitement de la date conso_train
conso_train["date"] = conso_train["date"].str.split('+').apply(lambda x:x[0])
```

```python
conso_train["date_only"] = conso_train["date"].str.split('T').apply(lambda x:x[0])
conso_train["heure_only"] = conso_train["date"].str.split('T').apply(lambda x:x[1])                                                                
```

```python
conso_train["full_date"] = conso_train["date"].str.replace('T',' ')
conso_train.drop('date',axis=1)
```

```python
conso_train["full_date"] = pd.to_datetime(conso_train["full_date"], format="%Y-%m-%d %H:%M:%S")
```

```python
conso_train.set_index('full_date', inplace=True)
conso_train.head()
```

```python
power = conso_train['2015-09-13':'2016-09-13'].groupby(pd.Grouper(freq='3H')).sum()
```

```python
# Drop l'index et le remplace par un index numérique par défaut
power = power.reset_index()
```

```python
len(power)
```

```python
train = pd.concat([meteo_train, power], axis=1)
```

```python
del train['Date UTC']
del train['full_date']
```
